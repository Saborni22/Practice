#### Non-blind SMC for TARN. Manage to get it to work pretty okay.

### Description: 

* Let $g(S_{1:m}) = \sum_{i=1}^{\tau} f(S_{i})$. Then $g_{m} \geq 0$ and final payoff is of the form 
$$ H(S_{1:m}) = g_{m}(S_{1:m}) \mathbb{1} \{ L_{5} < 100 \}$$ 
* We want to try weighting functions:
$$ H(S_{1:m}) = h_{1}(S_{1}) \times \frac{ h_{2}(S_{1:2}) }{ h_{1} (S_{1}) } \times \ldots \times \frac{ h_{5}(S_{1:5}) }{ h_{4} (S_{1:4}) } \times \frac{ \mathbb{1} \{ L_{5} < 100 \}} { h_{5}(S_{1:5}) } \times g_{m}(S_{1:m}) $$
* We play around a bit with the weighting functions.

```
# Calling libraries:
%matplotlib inline  
from __future__ import division
import numpy as np
import time
from scipy.stats import norm, uniform
from scipy.interpolate import interp1d
from pylab import plot, show, legend
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (15.0, 5.0) #for larger plots
int = np.vectorize(int)
```

```
# Payoff function:
def f(x):
    s = np.exp(x)
    return (2*(s-110)+20)*(s>110) + (2*(80-s)+20)*(s<90) - 20*(s<=110)*(s>=90)  
def f_positive(x):
    return f(x)*(f(x)>0)
def f_negative(x):
    return -f(x)*(f(x)<0)
def payoff_function(x):
    return 100 + f(x)
np.vectorize(f)
np.vectorize(f_positive)
np.vectorize(f_negative)
np.vectorize(payoff_function)
```

```
# Plotting payoff function:

x = np.log(np.linspace(start=60,stop=140,num=1000))
plt.rcParams['figure.figsize'] = (15.0, 5.0)
plt.figure()
plt.xlabel('$S$')
plt.ylabel('$f(S)$')
plt.title('Payoff Function $f$')
plt.xlim(60,140)
plt.ylim(-25,80)
plt.plot(np.exp(x),f(x),'b-')
plt.plot([100,100],[-25,80],'g-')
plt.plot([60,140],[0,0],'g-')
plt.grid(True)
plt.show()
```

```
# Systematic resampling function: 
def systematic_resampling(population, weights, index_resampled):
    weights /= np.sum(weights)
    population_length = len(population)
    u = np.random.random() / population_length
    p_index=0; sum_weights = 0
    for k in xrange(population_length):
        sum_weights += weights[k]
        while (u < sum_weights ):
            index_resampled[p_index] = k
            p_index+=1
            u = u + 1.0/population_length
```

### Simple Black-Scholes Dynamics (constant volatility):

```
# Initialization:
m, k, M = 24, 30, 10**5
gamma_L, gamma_G = 100, 200
S_0, mu, sigma = np.log(100), 0, 0.05
rep = 100
```

```
# Naive Monte Carlo for TARN price:

def MC_TARN(m,k,S_0,sigma,mu,M,gamma_G,gamma_L):
    S, G, L = [S_0]*M, np.zeros(M), np.zeros(M)
    price, survive = np.zeros(M), np.ones(M)
    mean, sd = (mu-0.5*sigma**2)*k/360, sigma*np.sqrt(k/360)
    for n in range(m):
        S += mean + sd*norm.rvs(0,1,M)
        price += f(S)*survive
        G += f_positive(S)*survive
        L += f_negative(S)*survive
        survive *= (G<gamma_G)*(L<gamma_L)     # these are the ones which are still 'alive' after time n
    return np.mean(price) + 100
```

```
# Non-blind SMC: 

def SMC_TARN_nb(m,k,S_0,sigma,mu,M,gamma_G,gamma_L):
    
    S, G, L = [S_0]*M, np.zeros(M), np.zeros(M)
    weights, normalizing_constant = [1/M]*M, 1.0
    survive, payoff_S, index_resampled = np.ones(M), np.zeros(M), int(np.ones(M))
    mean, sd = (mu-0.5*sigma**2)*k/360, sigma*np.sqrt(k/360)
    
    # Time steps 1 to 5:
    for n in np.add(1,range(5)):
        weights /= weighting_function(S,n-1)
        S += mean + sd*norm.rvs(0,1,M)
        weights *= weighting_function(S,n)
        normalizing_constant *= np.sum(weights)
        systematic_resampling(S,weights,index_resampled) 
        S, payoff_S, G, L = S[index_resampled], payoff_S[index_resampled], G[index_resampled], L[index_resampled]
        survive, weights = survive[index_resampled], [1/M]*M
        payoff_S += f(S)*survive
        G += f_positive(S)*survive
        L += f_negative(S)*survive
        survive *= (G<gamma_G)*(L<gamma_L)
    payoff_S += 100 
    
    # Time step 6:
    weights *= (L<gamma_L)/weighting_function(S,5)
    n_escaped = np.sum(L<gamma_L)
    normalizing_constant *= np.sum(weights)
    weights /= np.sum(weights)
    systematic_resampling(S,weights,index_resampled) 
    S, payoff_S, G, L = S[index_resampled], payoff_S[index_resampled], G[index_resampled], L[index_resampled]
    survive, weights = survive[index_resampled], [1/M]*M
    
    # Time steps 7 to 24 
    for n in range(m-6):
        S += mean + sd*norm.rvs(0,1,M)
        payoff_S += f(S)*survive
        G += f_positive(S)*survive
        L += f_negative(S)*survive
        survive *= (G<gamma_G)*(L<gamma_L) 
    weights *= payoff_S
    normalizing_constant *= np.sum(weights)
    weights /= np.sum(weights)
    
    return normalizing_constant, n_escaped
```

### Variance comparison:

```
def compare_methods(rep,M,sigma):
    MC_TARN_estimate, SMC_TARN_estimate = np.zeros(rep), np.zeros(rep)
    start_time = time.clock()
    for r in range(rep):
        MC_TARN_estimate[r] = MC_TARN(m,k,S_0,sigma,mu,M,gamma_G,gamma_L)
    MC_time = time.clock() - start_time 
    start_time = time.clock()
    for r in range(rep):
        SMC_TARN_estimate[r] = SMC_TARN_nb(m,k,S_0,sigma,mu,M,gamma_G,gamma_L)[0]
    SMC_time = time.clock() - start_time 
    print "sigma =", sigma, ", no. of particles =", M
    print "Mean of MC, SMC =", round(np.mean(MC_TARN_estimate),4), ",", round(np.mean(SMC_TARN_estimate),4)
    print "Average time for one run of MC, SMC =", round(MC_time/rep,2), ",", round(SMC_time/rep,2), "sec."
    print "Relative standard deviation of MC to SMC =", round(np.std(MC_TARN_estimate)/np.std(SMC_TARN_estimate),2)
    print "Total time for", rep, "repetitions =", round((MC_time+SMC_time)/60,2), "minutes"
```

```
def plot_results(rep,M,sigma_list):
    plt.rcParams['figure.figsize'] = (15.0, 5.0)
    n_sigma = len(sigma_list)
    MC_price, SMC_price = np.zeros(rep), np.zeros(rep)
    rel_std = np.zeros(n_sigma)
    start_time = time.clock()
    for n in range(n_sigma):
        for r in range(rep):
            MC_price[r] = MC_TARN(m,k,S_0,sigma_list[n],mu,M,gamma_G,gamma_L)
            SMC_price[r] = SMC_TARN_nb(m,k,S_0,sigma_list[n],mu,M,gamma_G,gamma_L)[0]
        rel_std[n] = np.std(MC_price)/np.std(SMC_price)
    
    print "No. of particles =", M, ", no. of reps =", rep, ", total time =", round((time.clock()-start_time)/60,2), "mins"
    
    x_min, x_max = np.min(sigma_list)-0.1*(np.max(sigma_list)-np.min(sigma_list)), np.max(sigma_list)+0.1*(np.max(sigma_list)-np.min(sigma_list))
    plt.plot(sigma_list,rel_std,'bo')
    plt.title("Relative standard deviation of MC over SMC")
    plt.xlabel("Volatility")
    plt.xlim(x_min,x_max)
    plt.xticks(sigma_list)
    plt.ylim(-1,1.2*np.max(rel_std))
    plt.ylabel("Relative standard deviation")
    plt.grid(True)
    plt.show()
    
    return sigma_list, rel_std
```

#### A slightly more sophisticated weighting function:

```
def weighting_function(S,n):
    M = len(S)
    if(n==0):
        weight = np.ones(M)
    if(n>0):
        mean, sd = S_0 + (mu-0.5*sigma**2)*n*k/360, sigma*np.sqrt(n*k/360)
        marginal_density = np.exp(-(S-mean)**2/(2*sd**2))
        target = np.abs(S-S_0)**2
        weight = target/marginal_density
    return weight
```
