#### (This uses approximating functions and SMC samplers to try to price TARN).

### Description: 
* Refer to TARN_3rd_attempt_SMC_samplers.
* Working for the particular example there.
* Consider a sequence of approximating functions $\{ f_{n} \}_{1 \leq n \leq p-1}$ and $f_{p} \equiv f$, $f_{0} \equiv 1/m$.
* Going to use SMC samplers, by setting 
$$\tilde{\pi}_{0}(s_{1:m}) = p(s_{1:m})$$ 
and 
$$\tilde{\pi}_{n}(s_{1:m}) = \left | \sum_{i=1}^{\tau_{f_{n}}} f_{n}(s_{i}) \right | p(s_{1:m}) $$
* In that case, 
$$\tilde{\pi}_{p}(s_{1:m}) = \left | \sum_{i=1}^{\tau_{f}} f(s_{i}) \right | p(s_{1:m}) $$
* Since $m$ if fixed, we drop it from $\tau_{m}$. The stopping times now also depend on $f_{n}$.
* $\tau_{g}$ denotes the stopping time with respect to a function $g$.

```python
# Calling libraries:
%matplotlib inline  
from __future__ import division
import numpy as np
import time
from scipy.stats import norm, uniform
from pylab import plot, show, legend
import matplotlib.pyplot as plt
import functools
plt.rcParams['figure.figsize'] = (10.0, 3.0) #for larger plots
int = np.vectorize(int)

# Initialization:
m, k, M = 4, 30, 10**4
gamma_L, gamma_G = 20, 35
S_0, mu, sigma = np.log(100), 0, 8
c_SMC, p = 0.5, 5
delta = np.linspace(start=4.5,stop=0,num=p)

def f(x):
    s = np.exp(x)
    return (2*(s-110)+20)*(s>110) + (2*(80-s)+20)*(s<90) - 20*(s<=110)*(s>=90)  
def f_positive(x):
    return f(x)*(f(x)>0)
def f_negative(x):
    return -f(x)*(f(x)<0)
np.vectorize(f)
np.vectorize(f_positive)
np.vectorize(f_negative)
```

#### Naive MC:

```python
# Naive Monte Carlo for TARN price:
def MC_TARN(m,k,S_0,sigma,mu,M,gamma_G,gamma_L):
    S, G, L = [S_0]*M, np.zeros(M), np.zeros(M)
    price, survive = np.zeros(M), np.ones(M)
    mean, sd = (mu-0.5*sigma**2)*k/360, sigma*np.sqrt(k/360)
    for n in range(m):
        S += mean + sd*norm.rvs(0,1,M)
        price += f(S)*survive
        G += f_positive(S)*survive
        L += f_negative(S)*survive
        survive *= (G<gamma_G)*(L<gamma_L)     # these are the ones which are still 'alive' after time n
    return np.mean(np.absolute(price))
```

#### Incremental weights for SMC:
* Incremental weights are:
$$ W_{n}^{(l)} \left ( x_{n-1}^{(l)} \right ) = \frac{ \tilde{\pi}_{n} \left( x_{n-1}^{(l)} \right ) }{ \tilde{\pi}_{n-1} \left ( x_{n-1}^{(l)} \right ) } = \frac{ \left | \sum_{i=1}^{\tau_{f_{n}}^{(l)}} f_{n} \left ( s_{i,n-1}^{(l)} \right ) \right |} { \left | \sum_{i=1}^{\tau_{f_{n-1}}^{(l)}} f_{n-1}\left ( s_{i,n-1}^{(l)} \right ) \right | }, \text{ } n \geq 1, $$
where $ x_{n}^{(l)} = s_{1:m,n}^{(l)}$, $n = 1, 2, \ldots, p$ and $l = 1, 2, \ldots, M$.
* $\tau^{(l)}_{f_{n}}$ denotes the stopping time of $s_{1:m,n-1}^{(l)}$ with respect to $f_{n}$ and $\tau^{(l)}_{f_{n-1}}$ denotes the stopping time of $s_{1:m,n-1}^{(l)}$ (same path) with respect to $f_{n-1}$.

```python
def f_approximate(x,d):
    m = np.size(x)
    s = np.exp(x)
    if(d!=0):
        v1 = (2*(80-s)+20)*(s<90) 
        v2 = -(20*(s-90)/d)*(90<=s)*(s<90+d)
        v3 = -20*(90+d<=s)*(s<=110-2*d)
        v4 = (20*(s-110)/d+20)*(110-2*d<s)*(s<=110)
        v5 = (2*(s-110)+20)*(110<s) 
        value = v1+v2+v3+v4+v5
    if(d==0):
        value = f(x)
    return(value)
np.vectorize(f_approximate)

def make_h(c):
    def h(x):
        return f_approximate(x,c)
    return h
```

#### Plotting approximating functions:

```python
x = np.log(np.linspace(start=75,stop=125,num=1000))
p = 10
delta = np.linspace(start=4.8,stop=0,num=p) 
plt.rcParams['figure.figsize'] = (10.0, 5.0)
plt.figure()
plt.xlabel('$s$')
plt.ylabel('$f_{n}(s)$')
plt.title('Approximating Functions')
for n in range(len(delta)):
    plt.plot(np.exp(x),f_approximate(x,delta[n]))
plt.grid(True)
plt.show()
```

#### Defining the SMC function: 

```python
# This is a function that gives the stopping time of a path 
# with respect to a function g and barriers gamma_L and gamma_G
# (assume that the path does not contain S_0 and is of length m)

def stopping_time(path,g,gamma_L,gamma_G):
    G, L, tau, survive = 0, 0, 1, 1
    G, L = np.asarray(G), np.asarray(L)
    m = np.size(path)
    while((survive==1)*(tau<m)==1):
        G = G + g(path[tau-1])*(g(path[tau-1])>0)
        L = L - g(path[tau-1])*(g(path[tau-1])<0)
        survive *= (G<gamma_G)*(L<gamma_L)
        tau += survive
    return tau
```

```python
# p(s_{1:m}): 
# This is the density of the path of a Brownian Motion starting at S_0
# (Brownian motion under our dynamics)
# 'path' does not contain S_0 and is of length m
# We consider the log of the density

def log_path_density(sigma,mu,s_0,path):
    path = np.append(s_0,path)
    N = np.size(path)
    mean = path[0:(N-1)] + (mu-0.5*sigma**2)*k/360.0
    sd = sigma*np.sqrt(k/360.0)
    return np.sum(np.log(norm.pdf(path[1:N],loc=mean,scale=sd)))
np.vectorize(log_path_density)
```

```python
# Symmetric random walk Metropolis-Hastings kernel:
# (No new array created)
# Also returns the stopping time for the propagated particle
# 'original' is of length m+1 an it does not contin S_0

def propagate(original,proposed,tau_original,g):
    proposed = norm.rvs(loc=original,scale=1.0) 
    tau_proposed = stopping_time(proposed,g,gamma_L,gamma_G)
    a = np.absolute(np.sum(g(proposed[0:tau_proposed]))/np.sum(g(original[0:tau_original])))
    b = np.exp(log_path_density(sigma,mu,S_0,proposed)-log_path_density(sigma,mu,S_0,original))
    MH_ratio = a*b
    acceptance_probability, accept = 1, 0
    if(MH_ratio<1.0):
        acceptance_probability = MH_ratio
    u = uniform.rvs(size=1)
    if(u<acceptance_probability):
        original, tau_original, accept = proposed, tau_proposed, 1
    return original, tau_original, accept
```

```python
# Defining the SMC function: (working on path-space)

def SMC_TARN(m,k,S_0,sigma,mu,M,c_SMC,p,gamma_L,gamma_G,delta):
    
    cut_off, normalizing_constant = c_SMC*M, 1.0
    S, proposed = np.zeros((M,m+1)), np.ones(M)
    S[:,0], weights = [S_0]*M, [1/M]*M
    mean, sd = (mu-0.5*sigma**2)*k/360, sigma*np.sqrt(k/360)
    accept, acceptance_rate, ESS = np.zeros(M), np.zeros(p), np.zeros(p)
    
    for n in np.add(1,range(m)):
        S[:,n] = S[:,n-1] + mean + sd*norm.rvs(0,1,M)
    S = S[:,1:m+1]
        
    # First time step:
    g = make_h(delta[0])
    for l in range(M):                                                              
        tau = stopping_time(S[l,:],g,gamma_L,gamma_G)
        weights[l] *= np.abs(np.sum(g(S[l,0:tau])))
        S[l,:], tau, accept[l] = propagate(S[l,:],proposed,tau,g)
    acceptance_rate[0] = np.mean(accept)
    normalizing_constant *= np.sum(weights) 
    weights /= np.sum(weights)
    ESS[0] = 1/np.sum(weights**2)
    if(ESS[0]<cut_off):
        S = S[np.random.choice(a=range(M),size=M,p=weights),:]
        weights = [1/M]*M 
    
    # Subsequent time steps:
    for n in np.add(2,range(p-1)):                # n = 2, 3, ..., p
        g_1, g_2 = make_h(delta[n-2]), make_h(delta[n-1])
        for l in range(M):                                                              
            tau = stopping_time(S[l,:],g_1,gamma_L,gamma_G)
            weights[l] /= np.abs(np.sum(g_1(S[l,0:tau])))
            tau = stopping_time(S[l,:],g_2,gamma_L,gamma_G)
            weights[l] *= np.abs(np.sum(g_2(S[l,0:tau])))
            S[l,:], tau, accept[l] = propagate(S[l,:],proposed,tau,g_2)
        acceptance_rate[n-1] = np.mean(accept)
        normalizing_constant *= np.sum(weights) 
        weights /= np.sum(weights)
        ESS[n-1] = 1/np.sum(weights**2)
        if(ESS[n-1]<cut_off):
            S = S[np.random.choice(a=range(M),size=M,p=weights),:]
            weights = [1/M]*M 
        
    return normalizing_constant, acceptance_rate, ESS
```

```python
MC_price = MC_TARN(m,k,S_0,sigma,mu,M,gamma_G,gamma_L)
SMC_price, acceptance_rate, ESS = SMC_TARN(m,k,S_0,sigma,mu,M,c_SMC,p,gamma_L,gamma_G,delta)

plt.subplot(2,1,1)
plt.xlabel('Effective Sample Size')
plt.ylabel('ESS')
plt.xlabel('$p$')
plt.plot(np.add(1,range(p)),ESS)
plt.ylim(0,1.1*M)
plt.xlim(0,p+1)
plt.grid(True)
plt.subplot(2,1,2)
plt.xlabel('Averge Acceptance Rate')
plt.ylabel('Acceptance Rate')
plt.xlabel('$p$')
plt.plot(acceptance_rate)
plt.ylim(0,1)
plt.xlim(0,p+1)
plt.grid(True)
plt.show()

print round(MC_price,1), ",", round(SMC_price,1)
```

* Code appears to work.
* Observe that the ESS does not change gradually. So I don't expect this to do any better, but let us see.

#### Comparing variance:

```python
def compare_methods(m,k,S_0,sigma,mu,M,gamma_G,gamma_L,rep,delta):
    MC_estimate, SMC_estimate = np.zeros(rep), np.zeros(rep)
    start_time = time.clock()
    for r in range(rep):
        MC_estimate[r] = MC_TARN(m,k,S_0,sigma,mu,M,gamma_G,gamma_L)
    MC_time = time.clock() - start_time 
    start_time = time.clock()
    for r in range(rep):
        SMC_estimate[r] = SMC_TARN(m,k,S_0,sigma,mu,M,c_SMC,p,gamma_L,gamma_G,delta)[0]
    SMC_time = time.clock() - start_time 
    print "m =", m, ", k =", k, ", M =", M, ", S_0 =", np.exp(S_0), ", sigma =", sigma, ", mu =", mu, ", gamma_G =", gamma_G, ", gamma_L =", gamma_L
    print "p =", p, ", delta =", delta
    print "Average time for one run of MC, SMC =", round(MC_time/rep,3), ",", round(SMC_time/rep,3), "sec."
    print "Mean of MC, SMC =", round(np.mean(MC_estimate),2), "," , round(np.mean(SMC_estimate),2)
    print "Relative standard deviation of MC to SMC =", round(np.std(MC_estimate)/np.std(SMC_estimate),2)
    print "Total time for", rep, "repetitions =", round((MC_time+SMC_time)/60,1), "minutes"
```

```python
rep, M, p = 50, 10**4, 10
delta = np.linspace(start=4.8,stop=0,num=p) 
compare_methods(m,k,S_0,sigma,mu,M,gamma_G,gamma_L,rep,delta)
```

* Not doing any better.
