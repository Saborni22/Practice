### Computing sensitivities (naively) for TARN: 
* We look at sensitivities for TARN to see how it behaves.
* Doing this naively, but it should suffice to give an idea.
* Compute sensitivies with respect to $S_{0}, \sigma, \Gamma_{L}, \Gamma_{G}.$

```python
# Calling libraries:
%matplotlib inline  
from __future__ import division
import numpy as np
import time
from scipy.stats import norm, uniform
from pylab import plot, show, legend
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (16.0, 5.0) #for larger plots
```

```python
def f(x):
    s = np.exp(x)
    return (2*(s-110)+20)*(s>110) + (2*(80-s)+20)*(s<90) - 20*(s<=110)*(s>=90)  

def f_positive(x):
    return f(x)*(f(x)>0)

def f_negative(x):
    return -f(x)*(f(x)<0)

np.vectorize(f)
np.vectorize(f_positive)
np.vectorize(f_negative)
```

```python
# Naive Monte Carlo for TARN price:

def naive_MC_TARN(m,k,S_0,sigma,mu,M,gamma_G,gamma_L):
    
    S, G, L = [S_0]*M, np.zeros(M), np.zeros(M)
    price, survive = np.zeros(M), np.ones(M)
    mean, sd = (mu-0.5*sigma**2)*k/360, sigma*np.sqrt(k/360)
    
    for n in range(m):
        S += mean + sd*norm.rvs(0,1,M)
        price += f(S)*survive
        G += f_positive(S)*survive
        L += f_negative(S)*survive
        survive *= (G<gamma_G)*(L<gamma_L)     # these are the ones which are still 'alive' after time n
    
    return np.mean(np.absolute(price))
```

```python
m, k, M = 24, 30, 10**5
gamma_L, gamma_G = 100, 200
S_0, mu, sigma = np.log(100), 0, 8
```

```python
# Computing sensitivities naively
# doing it on a grid of 500 points

d = 501             # number of grid points
def sensitivities_naive(m,k,M,d):
    gamma_L_grid = np.linspace(start=50,stop=150,num=d)
    gamma_G_grid = np.linspace(start=150,stop=250,num=d)
    sigma_grid = np.linspace(start=4,stop=12,num=d)
    S_0_grid = np.linspace(start=np.log(70),stop=np.log(130),num=d)
    
    price_gamma_L, price_gamma_G, price_sigma, price_S_0 = np.zeros(d), np.zeros(d), np.zeros(d), np.zeros(d)
    
    for r in range(d):
        price_gamma_L[r] = naive_MC_TARN(m,k,S_0,sigma,mu,M,gamma_G,gamma_L_grid[r])
        price_gamma_G[r] = naive_MC_TARN(m,k,S_0,sigma,mu,M,gamma_G_grid[r],gamma_L)
        price_sigma[r] = naive_MC_TARN(m,k,S_0,sigma_grid[r],mu,M,gamma_G,gamma_L)
        price_S_0[r] = naive_MC_TARN(m,k,S_0_grid[r],sigma,mu,M,gamma_G,gamma_L)

    # Plotting the sensitivities
    plt.figure()
    plt.plot(gamma_L_grid[0:d-1],(price_gamma_L[1:d]-price_gamma_L[0:d-1])/(100/d))
    plt.xlabel('Gamma_L')
    plt.ylabel('Sensitivity')
    plt.title('Sensitivity of price to Gamma_L')
    plt.grid(True)
    plt.show()
    
    plt.figure()
    plt.plot(gamma_G_grid[0:d-1],(price_gamma_G[1:d]-price_gamma_G[0:d-1])/(100/d))
    plt.xlabel('Gamma_G')
    plt.ylabel('Sensitivity')
    plt.title('Sensitivity of price to Gamma_G')
    plt.grid(True)
    plt.show()
    
    plt.figure()
    plt.plot(sigma_grid[0:d-1],(price_sigma[1:d]-price_sigma[0:d-1])/(8/d))
    plt.xlabel('Volatility')
    plt.ylabel('Sensitivity')
    plt.title('Sensitivity of price to Volatlity')
    plt.grid(True)
    plt.show()
    
    plt.figure()
    plt.plot(np.exp(S_0_grid[0:d-1]),(price_S_0[1:d]-price_S_0[0:d-1])/(np.log(13/7)/d))
    plt.xlabel('Initial value')
    plt.ylabel('Sensitivity')
    plt.title('Sensitivity of price to Initial Value')
    plt.grid(True)
    plt.show()
```

```python
sensitivities_naive(24,30,10**6,501)
```
