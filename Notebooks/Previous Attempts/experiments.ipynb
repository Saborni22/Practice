#### Some experiments. Trying to get paths to move outside the pit by time 5.

```python
# Calling libraries:
%matplotlib inline  
from __future__ import division
import numpy as np
import time
from scipy.stats import norm, uniform
from pylab import plot, show, legend
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (16.0, 1.5) #for larger plots
int = np.vectorize(int)

# Initialization:
m, k, M = 24, 30, 10**5
gamma_L, gamma_G = 100, 200
S_0, mu, sigma = np.log(100), 0, 0.08
c_SIR = 0.5

# Payoff function:
def f(x):
    s = np.exp(x)
    return (2*(s-110)+20)*(s>110) + (2*(80-s)+20)*(s<90) - 20*(s<=110)*(s>=90)  
def f_positive(x):
    return f(x)*(f(x)>0)
def f_negative(x):
    return -f(x)*(f(x)<0)
np.vectorize(f)
np.vectorize(f_positive)
np.vectorize(f_negative)
```

```python
def payoff(path,gamma_G,gamma_L):
    m, n, value = len(path), 0, 0
    G, L, survive = 0, 0, 1
    while((survive!=0)*(n<m)==1):
        value += f(path[n])
        G += f_positive(path[n])
        L += f_negative(path[n])
        survive *= (G<gamma_G)*(L<gamma_L)
        n += 1
    return value
```

```python
def g(x):
    x *= 100
    return np.exp(x)
def h(x):
    x *= 100
    return x**2
np.vectorize(g)
```

```python
def weighting_function(S,S_0,n):
    if(n==0):
        value = np.ones(M)
    if((n>0)*(n<=5)==1):
        value = g(np.abs(S-S_0))
    if(n>5):
        value = h(np.abs(S-S_0))
    return value
```

```python
S, weights = [S_0]*M, [1/M]*M
cut_off, normalizing_constant, ESS = c_SIR*M, 1, np.zeros(m)
mean, sd = (mu-0.5*sigma**2)*k/360, sigma*np.sqrt(k/360)
numerator, denominator = np.zeros(M), np.zeros(M)
x_min, x_max = 50, 160
G, L, survive, f_S = np.zeros(M), np.zeros(M), np.ones(M), np.zeros(M)
    
# Time steps 1 to m-1
for n in np.add(1,range(m-1)):        # n = 1, 2, ..., m-1
    weights /= weighting_function(S,S_0,n-1)
    S += mean + sd*norm.rvs(0,1,M)
    weights *= weighting_function(S,S_0,n)
    normalizing_constant *= np.sum(weights)
    f_S += f(S)*survive
    G += f_positive(S)*survive
    L += f_negative(S)*survive
    survive *= (G<gamma_G)*(L<gamma_L)
    weights /= np.sum(weights)
    ESS[n-1] = 1/np.sum(weights**2)
    if(ESS[n-1]<cut_off):
        resampling_index = np.random.choice(a=range(M),size=M,p=weights)
        S, G, L = S[resampling_index], G[resampling_index], L[resampling_index] 
        weights, survive, f_S = [1/M]*M, survive[resampling_index], f_S[resampling_index]
    print "n =", n, ":::::"
    plt.hist(np.exp(S),bins=50,normed=True)
    plt.xlim(x_min, x_max)
    plt.title('Histogram of $S_{n}$')
    plt.grid(True)
    plt.show()

# Final time step:
weights /= weighting_function(S,S_0,m-1)
S += mean + sd*norm.rvs(0,1,M)
f_S += f(S)*survive
weights *= f_S
normalizing_constant *= np.sum(weights)
    
plt.plot(ESS)
plt.ylim(0,1.1*np.max(ESS))
plt.title('ESS')
plt.grid(True)

print normalizing_constant
```
