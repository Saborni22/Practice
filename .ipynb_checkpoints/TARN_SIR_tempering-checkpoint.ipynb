#### (This uses sequences of approximating functions and SIR to try to price TARNS).

### Description:
* Refer to TARN_2nd_attempt_SIR.
* Now focused on the particular payoff function described there.
* Notice that $f$ is discontinuous. 
* We consider a sequence of continuous functions $\{ f_{n} \}_{1 \leq n \leq m-1}$ converging to $f$ and $f_{m} \equiv f$ and use these are the weighting functions to do SIR (instead of using a temperature scheme).
* Refer to pdf file for more details.

```python
# Calling libraries:
%matplotlib inline  
from __future__ import division
import numpy as np
import time
from scipy.stats import norm, uniform
from pylab import plot, show, legend
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (10.0, 3.0) #for larger plots
int = np.vectorize(int)

# Initialization:
m, k, M = 24, 30, 10**4
gamma_L, gamma_G = 100, 200
S_0, mu, sigma = np.log(100), 0, 0.08
c_SIR = 0.5

# Payoff function:
def f(x):
    s = np.exp(x)
    return (2*(s-110)+20)*(s>110) + (2*(80-s)+20)*(s<90) - 20*(s<=110)*(s>=90)  
def f_positive(x):
    return f(x)*(f(x)>0)
def f_negative(x):
    return -f(x)*(f(x)<0)
np.vectorize(f)
np.vectorize(f_positive)
np.vectorize(f_negative)
```

#### Naive MC: 

```python
# Naive Monte Carlo for TARN price:
def MC_TARN(m,k,S_0,sigma,mu,M,gamma_G,gamma_L):
    S, G, L = [S_0]*M, np.zeros(M), np.zeros(M)
    price, survive = np.zeros(M), np.ones(M)
    mean, sd = (mu-0.5*sigma**2)*k/360, sigma*np.sqrt(k/360)
    for n in range(m):
        S += mean + sd*norm.rvs(0,1,M)
        price += f(S)*survive
        G += f_positive(S)*survive
        L += f_negative(S)*survive
        survive *= (G<gamma_G)*(L<gamma_L)     # these are the ones which are still 'alive' after time n
    return np.mean(np.absolute(price))
```

```python
def make_h(c):
    def h(x):
        return f_approximate(x,c)
    return h
```

### Stopping time w.r.t. final $f$:
* Incremental weights are
$$ \alpha_{n} \left (s_{1:n}^{(l)} \right ) = \frac{ \left | \sum_{i=1}^{\tau_{n,f}^{(l)}} f_{n} \left (s_{i}^{(l)} \right ) \right | } { \left | \sum_{i=1}^{\tau_{n-1,f}^{(l)}} f_{n-1} \left ( s_{i}^{(l)} \right) \right | } $$

```python
# SIR for TARN 
# multinomial resampling for the time being
# using process density as proposal 

def SIR_TARN_v1(m,k,S_0,sigma,mu,M,c,gamma_G,gamma_L,delta):
    
    S, weights, survive, tau = np.zeros((M,m)), [1/M]*M, np.ones(M), np.ones(M)
    cut_off, normalizing_constant, ESS = c*M, 1, np.zeros(m)
    mean, sd = (mu-0.5*sigma**2)*k/360, sigma*np.sqrt(k/360)
    numerator, denominator = np.zeros(M), np.zeros(M)
    
    # First month:
    S[:,0] = S_0 + mean + sd*norm.rvs(0,1,M)
    G, L = f_positive(S[:,0]), f_negative(S[:,0])
    weights *= np.abs(f_approximate(S[:,0],delta[0]))
    normalizing_constant *= np.sum(weights)
    weights /= np.sum(weights)
    ESS[0] = 1/np.sum(weights**2)
    if(ESS[0]<cut_off):
        resampling_index = np.random.choice(a=range(M),size=M,p=weights)
        S[:,0], G, L = S[resampling_index,0], G[resampling_index], L[resampling_index]
        weights = [1/M]*M
    survive = (G<gamma_G)*(L<gamma_L)
    tau += survive
        
    # Subsequent months:
    for n in np.add(1,range(m-1)):      # n = 1, 2, ..., m-1
        g1, g2 = make_h(delta[n-1]), make_h(delta[n])
        S[:,n] = S[:,n-1] + mean + sd*norm.rvs(0,1,M)
        for l in range(M):      # (how to avoid this loop?)
            numerator[l] = np.abs(np.sum(g2(S[l,0:tau[l]]))*(tau[l]<n+1)+np.sum(g2(S[l,0:n+1])*(tau[l]==n+1)))
            denominator[l] = np.abs(np.sum(g1(S[l,0:tau[l]]))*(tau[l]<n+1)+np.sum(g1(S[l,0:tau[l]-1]))*(tau[l]==n+1))
        weights *= numerator/denominator
        G += f_positive(S[:,n])*survive
        L += f_negative(S[:,n])*survive
        survive *= (G<gamma_G)*(L<gamma_L)
        tau += survive
        normalizing_constant *= np.sum(weights)
        weights /= np.sum(weights)
        ESS[n] = 1/np.sum(weights**2)
        if(ESS[n]<cut_off):
            resampling_index = np.random.choice(a=range(M),size=M,p=weights)
            S[:,0:n+1], G, L = S[resampling_index,0:n+1], G[resampling_index], L[resampling_index]
            survive, weights, tau = survive[resampling_index], [1/M]*M, tau[resampling_index]
                
    return normalizing_constant, ESS
```

#### Sequence of approximating functions $f_{n}$: 

```python
def f_approximate(x,d):
    s = np.exp(x)
    if(d!=0):
        v1 = (2*(80-s)+20)*(s<90) 
        v2 = -(20*(s-90)/d)*(90<=s)*(s<90+d)
        v3 = -20*(90+d<=s)*(s<=110-2*d)
        v4 = (20*(s-110)/d+20)*(110-2*d<s)*(s<=110)
        v5 = (2*(s-110)+20)*(110<s) 
        value = v1+v2+v3+v4+v5
    if(d==0):
        value = f(x)
    return value
np.vectorize(f_approximate)

delta = np.linspace(start=4.8,stop=0,num=m)
x = np.log(np.linspace(start=75,stop=125,num=1000))
plt.rcParams['figure.figsize'] = (10.0, 3.0)
plt.figure()
plt.xlabel('$s$')
plt.ylabel('$f_{n}(s)$')
plt.title('Approximating Functions')
for n in range(len(delta)):
    plt.plot(np.exp(x),f_approximate(x,delta[n]))
plt.grid(True)
plt.show()
```

```python
MC_price = MC_TARN(m,k,S_0,sigma,mu,M,gamma_G,gamma_L)
SIR_price, ESS = SIR_TARN_v1(m,k,S_0,sigma,mu,M,c_SIR,gamma_G,gamma_L,delta)

plt.figure()
plt.xlabel('Time')
plt.ylabel('ESS')
plt.title('Effective Sample Size for SIR')
plt.plot(ESS)
plt.ylim(0,1.1*M)
plt.grid(True)
plt.show()

print round(MC_price,1), ",", round(SIR_price,1)
```

### Variance comparison:

```python
def compare_methods(m,k,S_0,sigma,mu,M,c_SIR_TARN,gamma_G,gamma_L,rep,delta):
    MC_TARN_estimate, SIR_TARN_estimate = np.zeros(rep), np.zeros(rep)
    start_time = time.clock()
    for r in range(rep):
        MC_TARN_estimate[r] = MC_TARN(m,k,S_0,sigma,mu,M,gamma_G,gamma_L)
    MC_time = time.clock() - start_time 
    start_time = time.clock()
    for r in range(rep):
        SIR_TARN_estimate[r] = SIR_TARN_v1(m,k,S_0,sigma,mu,M,c_SIR_TARN,gamma_G,gamma_L,delta)[0]
    SIR_time = time.clock() - start_time 
    print "m =", m, ", k =", k, ", M =", M, ", S_0 =", np.exp(S_0), ", sigma =", sigma, ", mu =", mu, ", gamma_G =", gamma_G, ", gamma_L =", gamma_L
    print "Average time for one run of MC, SIR =", round(MC_time/rep,2), ",", round(SIR_time/rep,2), "sec."
    print "Mean of MC, SIR =", round(np.mean(MC_TARN_estimate),3), ",", round(np.mean(SIR_TARN_estimate),3)
    print "Standard deviation for MC, SIR =", round(np.std(MC_TARN_estimate),3), ",", round(np.std(SIR_TARN_estimate),3)
    print "Relative standard deviation of MC to SIR =", round(np.std(MC_TARN_estimate)/np.std(SIR_TARN_estimate),2)
    print "Total time for", rep, "repetitions =", round((MC_time+SIR_time)/60,2), "minutes"
```

```python
rep, M = 50, 10**4
compare_methods(m,k,S_0,sigma,mu,M,c_SIR,gamma_G,gamma_L,rep,delta)
```

### Stopping time w.r.t. each $f_{n}$:
* incremental weights are
$$ \alpha_{n} \left (s_{1:n}^{(l)} \right ) = \frac{ \left | \sum_{i=1}^{\tau_{n,f_{n}}^{(l)}} f_{n} \left (s_{i}^{(l)} \right ) \right | } { \left | \sum_{i=1}^{\tau_{n-1,f_{n-1}}^{(l)}} f_{n-1} \left ( s_{i}^{(l)} \right) \right | } $$

```python
# This is a function that gives the stopping time of a path 
# with respect to a function g and barriers gamma_L and gamma_G
# (assume that the path does not contain S_0 and is of length m)

def stopping_time(path,g,gamma_L,gamma_G):
    G, L, tau, survive = 0, 0, 1, 1
    G, L = np.asarray(G), np.asarray(L)
    m = np.size(path)
    while((survive==1)*(tau<m)==1):
        G = G + g(path[tau-1])*(g(path[tau-1])>0)
        L = L - g(path[tau-1])*(g(path[tau-1])<0)
        survive *= (G<gamma_G)*(L<gamma_L)
        tau += survive
    return tau
```

```python
# SIR for TARN 
# multinomial resampling for the time being
# using process density as proposal 

def SIR_TARN_v2(m,k,S_0,sigma,mu,M,c,gamma_G,gamma_L,delta):
    
    S, weights, survive, tau = np.zeros((M,m)), [1/M]*M, np.ones(M), np.ones(M)
    cut_off, normalizing_constant, ESS = c*M, 1, np.zeros(m)
    mean, sd = (mu-0.5*sigma**2)*k/360, sigma*np.sqrt(k/360)
    numerator, denominator = np.zeros(M), np.zeros(M)
    
    # First month:
    g1 = make_h(delta[0])
    S[:,0] = S_0 + mean + sd*norm.rvs(0,1,M)
    G, L = g1(S[:,0])*(g1(S[:,0])>0), -g1(S[:,0])*(g1(S[:,0])<0)
    weights *= np.abs(g1(S[:,0]))
    normalizing_constant *= np.sum(weights)
    weights /= np.sum(weights)
    ESS[0] = 1/np.sum(weights**2)
    if(ESS[0]<cut_off):
        resampling_index = np.random.choice(a=range(M),size=M,p=weights)
        S[:,0], G, L = S[resampling_index,0], G[resampling_index], L[resampling_index]
        weights = [1/M]*M
    survive = (G<gamma_G)*(L<gamma_L)
    tau += survive
        
    # Subsequent months:
    for n in np.add(1,range(m-1)):      # n = 1, 2, ..., m-1
        g1, g2 = make_h(delta[n-1]), make_h(delta[n])
        S[:,n] = S[:,n-1] + mean + sd*norm.rvs(0,1,M)
        for l in range(M):      # (how to avoid this loop?)
            denominator[l] = np.abs(np.sum(g1(S[l,0:tau[l]]))*(tau[l]<n+1)+np.sum(g1(S[l,0:tau[l]-1]))*(tau[l]==n+1))
            tau[l] = stopping_time(S[l,1:n+1],g2,gamma_L,gamma_G)
            numerator[l] = np.abs(np.sum(g2(S[l,0:tau[l]])))
        weights *= numerator/denominator
        normalizing_constant *= np.sum(weights)
        weights /= np.sum(weights)
        ESS[n] = 1/np.sum(weights**2)
        if(ESS[n]<cut_off):
            resampling_index = np.random.choice(a=range(M),size=M,p=weights)
            S[:,0:n+1], G, L = S[resampling_index,0:n+1], G[resampling_index], L[resampling_index]
            survive, weights, tau = survive[resampling_index], [1/M]*M, tau[resampling_index]
                
    return normalizing_constant, ESS
```

### Variance Comparison:

```python
def compare_methods_2(m,k,S_0,sigma,mu,M,c_SIR_TARN,gamma_G,gamma_L,rep,delta):
    MC_TARN_estimate, SIR_TARN_estimate = np.zeros(rep), np.zeros(rep)
    start_time = time.clock()
    for r in range(rep):
        MC_TARN_estimate[r] = MC_TARN(m,k,S_0,sigma,mu,M,gamma_G,gamma_L)
    MC_time = time.clock() - start_time 
    start_time = time.clock()
    for r in range(rep):
        SIR_TARN_estimate[r] = SIR_TARN_v2(m,k,S_0,sigma,mu,M,c_SIR_TARN,gamma_G,gamma_L,delta)[0]
    SIR_time = time.clock() - start_time 
    print "m =", m, ", k =", k, ", M =", M, ", S_0 =", np.exp(S_0), ", sigma =", sigma, ", mu =", mu, ", gamma_G =", gamma_G, ", gamma_L =", gamma_L
    print "Average time for one run of MC, SIR =", round(MC_time/rep,2), ",", round(SIR_time/rep,2), "sec."
    print "Mean of MC, SIR =", round(np.mean(MC_TARN_estimate),1), ",", round(np.mean(SIR_TARN_estimate),1)
    print "Relative standard deviation of MC to SIR =", round(np.std(MC_TARN_estimate)/np.std(SIR_TARN_estimate),2)
    print "Total time for", rep, "repetitions =", round((MC_time+SIR_time)/60,2), "minutes"
```
