#### Some experiments. Trying to get paths to move outside the pit by time 5.

```python
# Calling libraries:
%matplotlib inline  
from __future__ import division
import numpy as np
import time
from scipy.stats import norm, uniform
from pylab import plot, show, legend
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (16.0, 1.5) #for larger plots
int = np.vectorize(int)

# Initialization:
m, k, M = 24, 30, 10**5
gamma_L, gamma_G = 100, 200
S_0, mu, sigma = np.log(100), 0, 0.08
c_SIR = 0.5

# Payoff function:
def f(x):
    s = np.exp(x)
    return (2*(s-110)+20)*(s>110) + (2*(80-s)+20)*(s<90) - 20*(s<=110)*(s>=90)  
def f_positive(x):
    return f(x)*(f(x)>0)
def f_negative(x):
    return -f(x)*(f(x)<0)
np.vectorize(f)
np.vectorize(f_positive)
np.vectorize(f_negative)
```

```python
def payoff(path,gamma_G,gamma_L):
    m, n, value = len(path), 0, 0
    G, L, survive = 0, 0, 1
    while((survive!=0)*(n<m)==1):
        value += f(path[n])
        G += f_positive(path[n])
        L += f_negative(path[n])
        survive *= (G<gamma_G)*(L<gamma_L)
        n += 1
    return value
```

```python
def g(x):
    x = np.exp(x)
    return np.exp(np.exp(np.exp(x)))
np.vectorize(g)
```

```python
def weighting_function(S,S_0):
    M, n = len(S[:,0]), len(S[0,:])
    if(n<5):
        value = g(np.abs(S[:,n-1]-S_0))
    if(n>=5):
        value = g(np.abs(S[:,n-1]-S_0))**0.5
    return value
```

```python
S, weights = np.zeros((M,m)), [1/M]*M
cut_off, normalizing_constant, ESS = c_SIR*M, 1, np.zeros(m)
mean, sd = (mu-0.5*sigma**2)*k/360, sigma*np.sqrt(k/360)
numerator, denominator = np.zeros(M), np.zeros(M)

# First time step:
S[:,0] = S_0 + mean + sd*norm.rvs(0,1,M)
weights *= g(np.abs(S[:,0]-S_0))
normalizing_constant *= np.sum(weights)
weights /= np.sum(weights)
ESS[0] = 1/np.sum(weights**2)
if(ESS[0]<cut_off):
    resampling_index = np.random.choice(a=range(M),size=M,p=weights)
    S[:,0], weights = S[resampling_index,0], [1/M]*M
print "n =", 1, ":::::"
plt.hist(np.exp(S[:,0]),bins=50,normed=True)
plt.xlim(0,400)
plt.title('Histogram of $S_{n}$')
plt.show()

# Second time step:
S[:,1] = S[:,0] + mean + sd*norm.rvs(0,1,M)
weights *= weighting_function(S[:,0:2],S_0)/g(np.abs(S[:,0]-S_0))
normalizing_constant *= np.sum(weights)
weights /= np.sum(weights)
ESS[0] = 1/np.sum(weights**2)
if(ESS[0]<cut_off):
    resampling_index = np.random.choice(a=range(M),size=M,p=weights)
    S[:,0], weights = S[resampling_index,0], [1/M]*M
print "n =", 2, ":::::"
plt.hist(np.exp(S[:,1]),bins=50,normed=True)
plt.title('Histogram of $S_{n}$')
plt.xlim(0,400)
plt.show()
    
# Subsequent time steps:
for n in np.add(2,range(m-3)):        # n = 2, ..., m-2
    S[:,n] = S[:,n-1] + mean + sd*norm.rvs(0,1,M)
    weights *= weighting_function(S[:,0:n],S_0)/weighting_function(S[:,0:n-1],S_0)
    normalizing_constant *= np.sum(weights)
    weights /= np.sum(weights)
    ESS[n] = 1/np.sum(weights**2)
    if(ESS[n]<cut_off):
        resampling_index = np.random.choice(a=range(M),size=M,p=weights)
        S[:,0:n+1], weights = S[resampling_index,0:n+1], [1/M]*M
    print "n =", n+1, ":::::"
    plt.hist(np.exp(S[:,n]),bins=50,normed=True)
    plt.xlim(0,400)
    plt.title('Histogram of $S_{n}$')
    plt.show()

# Final time step:
S[:,m-1] = S[:,m-2] + mean + sd*norm.rvs(0,1,M)
weights /= weighting_function(S[:,0:m-1],S_0)
for i in range(M):
    weights[i] *= payoff(S[i,:],gamma_G,gamma_L)
normalizing_constant *= np.sum(weights)
    
plt.plot(ESS)
plt.ylim(0,1.1*M)
plt.title('ESS')

print normalizing_constant
```
